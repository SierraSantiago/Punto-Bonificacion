```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import triang
```

La produccion de un articulo se distribuye trianfgularmente entre (4500, 6000 y 9000) uds. La fraccion defectuosa cada dia, p, se distribuye segun la siguiente fdp.  f(p)= 99*(1-p)^(98) , 0<=p<=1
P(unidades defectuosas>150)

```python

a, b, c = 4500, 6000, 9000  
scale = c - a  
c_param = (b - a) / scale  

# Crear la distribución triangular
triangular_dist = triang(c_param, loc=a, scale=scale)

def sample_defect_fraction(size=1):
    u = np.random.uniform(0, 1, size)  # Generar valores uniformes entre 0 y 1
    return 1 - (1 - u) ** (1/99)  # Aplicar la inversa de la CDF
```

```python
# Simulación
num_samples = 10000
production_samples = triangular_dist.rvs(num_samples)
defect_fractions = sample_defect_fraction(num_samples)
defective_units = production_samples * defect_fractions
```

```python
prob_defective_gt_150 = np.mean(defective_units > 150)
print(f"Probabilidad de unidades defectuosas > 150: {prob_defective_gt_150:.4f}")
```
